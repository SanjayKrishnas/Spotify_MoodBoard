```python
from sklearn.model_selection import train_test_split 
  
X = df.get(['speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo'])
y = df['time_signature'] 
  
# using the train test split function 
# random state will change the split of the data

# double shuffle ???
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=104,test_size=0.2,shuffle=True) 
  
# printing out train and test sets 
  
print('X_train : ') 
print(X_train.head()) 
print('') 
print('X_test : ') 
print(X_test.head()) 
print('') 
print('y_train : ') 
print(y_train.head()) 
print('') 
print('y_test : ') 
print(y_test.head())
```
